### How is an image formed with camera?

The objects in real world reflect light after receiving it from light sources like sun, bulbs, etc. This refelcted light is incident on the lens of the camera and then hits an array of sensors, pixels. The electric charge produced by the sensors are converted to voltages. These voltages are sampled by digitizers to produce pixel values which can be processed by computers. 

The process stated above summarizes the process of creating images in a nutshell. We can solidify our understanding with a simple 1D pinhole camera. 

A pinhole camera is a very simplistic representation of a camera without lens. An image is formed on a translucent screen by blocking light on its opposite through a pinhole. What the pinhole does is that it blocks out light coming from the object and only allows light to pass from one  point in the object which makes it easier to create an image on the translucent side. 

Taking an example of a tree, when you don't have a pinhole, multiple light rays from every point on the tree can be incident on the screen. But if you put a pinhole in between, only limited amount of rays will be able to project on screen, creating an image. 

Coming to 1D, a pinhole camera becomes 1D when we reduce the pinhole to a single point and we use it to block the light. Practically, no power passes through such a pinhole and we will get a black screen. But for geometric pusrposed, it's easier to build a 1D pinhole camera. 

### Some definitions used in the notebook

**Center of projection:** The pinhole is also called the center of projection

**Image plane:** The screen where the image is formed.

**Focal distance:** The distance between center of projection and image plane.

**Optical axis:** Line through the center of projection that is perpendicular to image plane

The image below shows 1D pinhole camera that projects 2D to 1D

The light is entering the pinhole camera, C at center of projection from a 2D point (X,Z). Now, we have to find the x coordinate of the point in the image plane. 

Using the similarity of triangle property, we can find the value of x:

$$-\frac{x}{f}=\frac{X}{Z}$$

$$x=-f \frac{X}{Z}$$

We can also move the image plane in front of C to get a positive value of the x coordinate. 


## An example in python

Let's design a 1D pinhole camera at a focal length of 20 pixels and find the projection of points (10,20) on the image plane.

```python
def pinhole_1d(focal_length, X, Z):
    x=(focal_length*X)/Z
    print("The projection of point ({},{}) is {}".format(X,Z,x))
```

```python
pinhole_1d(20,10,20)
```
